# set methods

```python
# element adding method
# add(ele): it adds the element to set if element not in set,otherwise nothing
s={1,2,3,4,5,4,3}
print(s)
s.add(100)
s.add(5)
print(s)
```

```python
#element removing method
#remove(item): it remove the item if it exist in set otherwise,'key error'
# discard(item): it remove the item if it exist in set otherwise nothing
```

```python
s={1,2,3,4,5}
s.remove(5)
print(s)
#s.remove(100) key error
s.discard(3)
print(s)
s.discard(100)
print(s)
```

set operations

```python
#set operations
#union(set): it returns the all unique elements from all sets
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,2,8,9}
union=s1.union(s2,s3)
print(union)
res=s1|s2|s3
print(res)
```

```python
#intersection(set): it returns the comman elements from all sets
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,5,8,9}
intersection=s1.intersection(s2,s3)
print(intersection)
res=s1&s2&s3
print(res)55
```

```python
#differnce (set): it returns the comman elements those are not in another sets 
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,5,8,9}
intersection=s1.difference(s2,s3)
print(intersection)
res=s1-s2-s3
print(res)
print((s1|s2)-s3)
```

```python
# symmetric_diffrence(set): it returns the non comman elements from all sets
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,5,8,9}
sy_differ=s1.symmetric_difference(s2)
print(sy_differ)
res=s1^s2
print(res)
```

```python
#issubset(set)
#issuperset(set)
#isdisjoint(set)
s1={1,2,3}
s2={1,2,3,4}
a=s1.issubset(s2)
print(a)
b=s1.issuperset(s2)
print(b)
c=s1.isdisjoint(s2)
print(c)
s={1,2,3}
print(dir(s))
```
