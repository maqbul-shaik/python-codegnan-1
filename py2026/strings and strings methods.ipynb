## string methods
-case conversion
-upper(): it converts entire string into upper case
-lower(): it converts entire string into lower case
-title(): it converts the first letter of every word upper remain lower
-swapcase(): it converts upper case letter into lower and lower to upper
-capitalize(): it converts first character into upper and remain lower

```python
str="the python "
print(str.upper())
print(str.lower())
print(str.title())
print(str.swapcase())
print(str.capitalize())
```

-remove white space or pattern at ends of string
-strip(): by default it removes white space of both ends of string
-lstrip(): it removes the white space at the left side of sting
-rstrip():it removes the white space at the rightside of sting

```python
s="  code   "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())  code   
code
code   
  code
```

```python
s="@@@code@@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@")) 
```

```python
#split(): it splits the entire string into list of substring based on specified separate
#by default it splits at space
s="irshad is a hero"
print(s.split())
print(s.split("is"))
print(s.split("e"))
```

# substring index finding
- index(substring): if substring exists string, it returns the first occurence of substring positions,otherwiswe valueerror
- find(substring):  if substring exists string, it returns the first occurence of substring positions,otherwiswe valueerror
- 

```python
s="the sky is blue and ocean is blue"
print(s.find('is'))
print(s.find('z'))
print(s.find('is',10))# it starts checking from the 10th position
print(s.find('e',10))
```

```python
#count(substring): it returns the total non-overlapping occurance of substring
s="ABCBBCCb"
print(s.count("z"))
l=s.lower()
print(l)
print(l.count("b"))
u=s.upper()
print(u.count("A"))
```

```python
# replace(old_substring,new_substring):it replace old_substring into new sub_string
s="the sky is Blue and ocean is blue"
print(s.replace("blue","rd"))
print(s.replace("e","E"))
print(s.replace("e","E",2))

```

### string checking methods
-string check9ing method returns the boolean value as result 
-isupper(): it return True, if entire string is uppercase
-islower(): it return True, if entire string is lowercase
-isalpha(): it return True, if entire string is alphabets
-isdigit(): it return True, if entire string is digits
-isalnum(): it return True, if entire string is contains either alphabets and digits
-startswith(substring): it return True, if string is starts with substring
-endsswith(substring): it return True, if string is ends with substring



```python
print("CODE".isupper())
print("CODE".islower())
print("123234".isalpha())
print("123".isdigit())
print("123er".isdigit())
print("CODE123".isalnum())
print("123".isalnum())
print("CODE".startswith("c"))
print("CODE".endswith("e"))
```
