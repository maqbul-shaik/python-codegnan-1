```python
#index and count
# index(item) : if item ispresent in list it returns first occurance the index otherwise value error
#count(item): it return the count of total occurance of item
lst=[10,20,30,40,20,50,60]
ind=lst.index(20)
print(ind)
print(lst.count(20))
```

```python
# sorting items
#sort(reverese=false): by default it sorts the original  list in into ascending order.it reverse is true it sort the decending order
lst=[2,3,4,5,6,-6,-7,45]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
```
