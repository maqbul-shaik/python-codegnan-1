```python
# Dictionary values acess by using keys
# updating dictionary values
#adding new items to dictionary
details={"name":"babu","course":"python","batch":57,"age":20}
print(details['course'])
#print(details['location']) #key error location
# updating dictionary values
details['age']=details['age']+1
print(details['age'])
#adding new items to dictionary
details['location']="hyderabad"
print(details)
```

```python
### Dictionary methods 
```

```python
# get(key, default_value:(optional): it returns the value if key is exists in dictionary otherwise is default value
```

```python
#dictionary keys are unique
details={'age':21,"name":"babu","course":"python","batch":57,"age":20}
print(details['name'])
print(details['marks']) # this one raises key error
print(details.get('name')) 
print(details.get('marks'))#get method default it returns none if key is not exists in
print(details.get('marks','mark is not available'))# it returns default value because
```

```python
-update(new_dictionary): this method is update teh value if new_dictionary keys already exists in dictionary,it adds new items to dictionary
```

```python
details={'age':21,"name":"babu","course":"python","batch":57,"age":20}
details.update({"batch":58,"course":"java","location":"hyd"})
print(details)
```

-keys(): this method returns the list dictionary keys
-values(): this method returns the list dictionary keys
items(): this method return the list of tuple of key-value pairs

```python
details={'age':21,"name":"babu","course":"python","batch":57,"age":20}
key=(details.keys())
print(type(key))
print(key)
print(details.values())
print(details.items())
```

-popitem():it return and removes the last item as tuple
-pop(key): this removes and specified key item and returned the specified key value,otherwise keyerror

```python
details={'age':21,"name":"babu","course":"python","batch":57}
item=details.popitem()
print(item)
print(details)
val=details.pop('name')
print(val)
print(details)
```

```python
#membership
details={'age':21,"name":"babu","course":"python","batch":57,"age":20}
print("babu" in details)
print('name' in details)
```
