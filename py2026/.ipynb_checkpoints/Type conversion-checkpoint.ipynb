```python
# str in to int,float,bool,list,tuple,set
#if string value is number
s='123456789'
print(int(s))
print(float(s))
print(str(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# str in to int,float,bool,list,tuple,set
#if string value is number
s='123456.76779'
print(int(float(s)))
print(float(s))
print(str(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# LIST in to str,tuple,set
#if string value is number
s=[1,2,3,4,5]
print(str(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# tuple in to str,list,set
s=(1,2,3,4,5)
print(str(s))
print(bool(s))
print(list(s))
print(set(s))
```

```python
#dic into str,tuple,list
s={"name":'hlo','count':25}
print(str(s))
print(bool(s))
print(list(s))
print(set(s))
```

```python
# set into str,list,tuple
s={1,2,4,5}
print(str(s))
print(bool(s))
print(list(s))
```
