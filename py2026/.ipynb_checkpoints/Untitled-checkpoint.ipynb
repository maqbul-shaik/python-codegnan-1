# input formating

```python
# reading integer from user
num=int(input("enter the number:"))
print(type(num))
print(num)        
```

```python
# reading float from user
num=float(input("enter the number:"))
print(type(num))
print(num) 
```

```python
# complex reding from user
val=complex(int(input()),int(input()))
print(type(val))
print(val)
```
